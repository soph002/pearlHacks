```python
import pip
import pandas as pd
import numpy as np
from numpy import linalg
import math
from scipy.linalg import svd
import matplotlib.pyplot as plt

data=pd.read_csv("SoftballData.csv")
print(data)
m=np.array(data)
print(m)
columns=data.shape[1]
rows=data.shape[0]
A=m[:,1:columns]
A=A.astype(float)
print(A)
u,s,v=np.linalg.svd(A)
print(s)
print(u)
x=u[:,0]
y=u[:,1]
for i in range(0,16):
  print(i,math.dist(u[0,0:2],u[i,0:2]))
#Keagan Goldwait has the closest hitting statistics to the best college softball hitter
#of all time

annotations=m[0:rows-1,0]
for i, label in enumerate(annotations):
  plt.text(x[i]+.001,y[i]-.002,label)
plt.scatter(x,y,alpha=0.5)
plt.show()
```
